```python
import keras
from keras.layers import *
import matplotlib.pyplot as plt
import numpy as np
import keras.utils
from tensorflow.keras.preprocessing import image
```

```python
(x_train, y_train), (x_test, y_test) = keras.datasets.fashion_mnist.load_data()
```

```python
input_shape = (28, 28, 1)
```

```python
model_1 = keras.Sequential([
      keras.Input(shape=input_shape),
      Conv2D(32, padding = 'same', kernel_size = (3,3), activation = 'relu' ),
      MaxPooling2D( (2,2), strides = 2),
      # Conv2D(64, padding = 'same', kernel_size = (3,3), activation = 'relu' ),
      # MaxPooling2D( (2,2), strides = 2),
      Flatten(), # перевод в одномерный массив
      Dense(50,activation = 'relu'), # количество входных нейронов
      Dense(10, activation = 'softmax')   # количество выходных нейронов
      ] )
```

```python
model_1.compile(optimizer='adam',
             loss='categorical_crossentropy',
             metrics=['accuracy'])
```

```python
x_train = x_train / 255
x_test = x_test / 255
y_train_cat = keras.utils.to_categorical(y_train, 10)
y_test_cat = keras.utils.to_categorical(y_test, 10)
```

```python
%%time

history = model_1.fit(x_train, y_train_cat, batch_size=32, epochs=5,validation_data =(x_test, y_test_cat) )
```

```python
%%time

history = model_4.fit(x_train, y_train_cat, batch_size=32, epochs=10,validation_data =(x_test, y_test_cat) )
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Точность модели')
plt.ylabel('точность')
plt.xlabel('количество эпох')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Потери модели')
plt.ylabel('Потери')
plt.xlabel('количество эпох')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

Для улучшения работы нейронной сети можно попробовать следующие методы:

Использование более сложной архитектуры модели: Добавление большего числа слоев и/или увеличение числа нейронов в слоях может помочь модели извлекать более сложные признаки из данных.

Увеличение числа эпох обучения: Иногда модели требуется больше времени для обучения, поэтому можно увеличить число эпох.

Использование других оптимизаторов: Помимо Adam, можно попробовать другие оптимизаторы, такие как RMSprop или SGD, с настройкой их гиперпараметров.

Добавление регуляризации: Добавление регуляризации, такой как dropout или L2 регуляризация, может помочь в борьбе с переобучением.

Использование аугментации данных: Применение случайных трансформаций к изображениям во время обучения может помочь модели генерализовать лучше.

Настройка гиперпараметров: Изменение параметров, таких как размер пакета (batch size) или размер ядра свертки, также может повлиять на производительность модели.

```python
from keras.optimizers import SGD
from keras.layers import Dropout

# Увеличиваем количество слоёв
model_5 = keras.Sequential([
      keras.Input(shape=input_shape),
      Conv2D(32, padding='same', kernel_size=(3,3), activation='relu'),
      MaxPooling2D((2,2), strides=2),
      Conv2D(64, padding='same', kernel_size=(3,3), activation='relu'),
      MaxPooling2D((2,2), strides=2),
      Conv2D(128, padding='same', kernel_size=(3,3), activation='relu'),
      MaxPooling2D((2,2), strides=2),
      Flatten(),
      Dense(128, activation='relu'),
      Dense(64, activation='relu'),
      Dense(10, activation='softmax')
])

model_5.compile(optimizer='adam',
                loss='categorical_crossentropy',
                metrics=['accuracy'])
```

```python
# Модель с dropout регуляризацией
model_7 = keras.Sequential([
      keras.Input(shape=input_shape),
      Conv2D(32, padding='same', kernel_size=(3,3), activation='relu'),
      MaxPooling2D((2,2), strides=2),
      Conv2D(64, padding='same', kernel_size=(3,3), activation='relu'),
      MaxPooling2D((2,2), strides=2),
      Flatten(),
      Dense(128, activation='relu'),
      Dropout(0.5),  # Dropout layer to reduce overfitting
      Dense(10, activation='softmax')
])

# Использование SGD
sgd = SGD(learning_rate=0.01, momentum=0.9, nesterov=True)
model_7.compile(optimizer=sgd,
                loss='categorical_crossentropy',
                metrics=['accuracy'])
```

```python
# Увеличение эпох
%%time

history = model_7.fit(x_train, y_train_cat, batch_size=32, epochs=15, validation_data=(x_test, y_test_cat))
```

```python
# Аугментация данных
from keras.preprocessing.image import ImageDataGenerator

datagen = ImageDataGenerator(
    rotation_range=10,
    width_shift_range=0.1,
    height_shift_range=0.1,
    shear_range=0.1,
    zoom_range=0.1,
    horizontal_flip=True,
    fill_mode='nearest')

# Создаем модель
input_shape = (28, 28, 1)

model = keras.Sequential([
    keras.Input(shape=input_shape),
    Conv2D(32, padding='same', kernel_size=(3, 3), activation='relu'),
    MaxPooling2D((2, 2), strides=2),
    Conv2D(64, padding='same', kernel_size=(3, 3), activation='relu'),
    MaxPooling2D((2, 2), strides=2),
    Conv2D(128, padding='same', kernel_size=(3, 3), activation='relu'),
    MaxPooling2D((2, 2), strides=2),
    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(10, activation='softmax')
])

# Компилируем модель
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
# Обучение модели с аугментацией данных
%%time

history = model.fit(datagen.flow(x_train.reshape(-1, 28, 28, 1), y_train_cat, batch_size=32),
                    steps_per_epoch=len(x_train) / 32, epochs=20,
                    validation_data=(x_test.reshape(-1, 28, 28, 1), y_test_cat))
```
